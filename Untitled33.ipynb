```python

# 1. Using sympy
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Variable simb√≥lica
t = sp.Symbol('t', real=True)

# Usamos np.pi en formato float para evitar problemas con sympy
pi = float(np.pi)
x0_vals = [-2*pi, -pi, 0, pi, 2*pi]   # posiciones iniciales

# Definimos funciones tipo abanico: x(t) = x0 * exp(-t)
x_funcs = [x0 * sp.exp(-t) for x0 in x0_vals]

# Convertimos a funciones num√©ricas con numpy
x_funcs_num = [sp.lambdify(t, xf, 'numpy') for xf in x_funcs]

# Dominio de tiempo
t_vals = np.linspace(0, 2, 200)

# Graficar
plt.figure(figsize=(6,6))
for f in x_funcs_num:
    f_vec = np.vectorize(f)  # <- fuerza a aceptar arrays
    plt.plot(t_vals, f_vec(t_vals), 'k')

# L√≠neas horizontales de referencia
plt.axhline(0, color='k', linewidth=0.8)
plt.axhline(pi, color='k', linewidth=0.8)
plt.axhline(-pi, color='k', linewidth=0.8)
plt.axhline(2*pi, color='k', linewidth=0.8)
plt.axhline(-2*pi, color='k', linewidth=0.8)

# L√≠nea vertical en t=0
plt.axvline(0, color='k', linewidth=0.8)

# Etiquetas
plt.xlabel("t")
plt.ylabel("x")
plt.title("Curvas caracter√≠sticas con Sympy + Matplotlib")
plt.grid(False)
plt.show()
```

```python
# 2. Using Euler
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n del m√©todo de Euler
def euler(f, x0, t0, tf, dt):
    """
    f : funci√≥n derivada dx/dt = f(x,t)
    x0: condici√≥n inicial
    t0: tiempo inicial
    tf: tiempo final
    dt: paso
    """
    N = int((tf - t0) / dt) + 1
    t = np.linspace(t0, tf, N)
    x = np.zeros(N)
    x[0] = x0
    for i in range(N - 1):
        x[i+1] = x[i] + dt * f(x[i], t[i])
    return t, x

# Definimos la EDO dx/dt = -x/(1+t)
def f(x, t):
    return -x / (1 + t)

# Condiciones iniciales
x0_vals = [-2*np.pi, -np.pi, 0, np.pi, 2*np.pi]

# Par√°metros de simulaci√≥n
t0, tf, dt = 0, 10, 0.01

# Crear figura
plt.figure(figsize=(6,6))

for x0 in x0_vals:
    t, x = euler(f, x0, t0, tf, dt)
    plt.plot(t, x, 'k')   # curva negra

# L√≠neas horizontales (gu√≠as)
for k in range(-2, 3):
    plt.axhline(k*np.pi, color='gray', linestyle='--', linewidth=0.7)

# Estilo del gr√°fico
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("M√©todo de Euler para dx/dt = -x/(1+t)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# 3. Using Ruge-Kutta
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n del m√©todo de Runge-Kutta 4
def rk4(f, x0, t0, tf, dt):
    """
    f : funci√≥n derivada dx/dt = f(x,t)
    x0: condici√≥n inicial
    t0: tiempo inicial
    tf: tiempo final
    dt: paso
    """
    N = int((tf - t0) / dt) + 1
    t = np.linspace(t0, tf, N)
    x = np.zeros(N)
    x[0] = x0
    for i in range(N - 1):
        k1 = f(x[i], t[i])
        k2 = f(x[i] + 0.5*dt*k1, t[i] + 0.5*dt)
        k3 = f(x[i] + 0.5*dt*k2, t[i] + 0.5*dt)
        k4 = f(x[i] + dt*k3, t[i] + dt)
        x[i+1] = x[i] + (dt/6)*(k1 + 2*k2 + 2*k3 + k4)
    return t, x

# Definimos la EDO dx/dt = -x/(1+t)
def f(x, t):
    return -x / (1 + t)

# Condiciones iniciales
x0_vals = [-2*np.pi, -np.pi, 0, np.pi, 2*np.pi]

# Par√°metros de simulaci√≥n
t0, tf, dt = 0, 10, 0.01

# Crear figura
plt.figure(figsize=(6,6))

for x0 in x0_vals:
    t, x = rk4(f, x0, t0, tf, dt)
    plt.plot(t, x, 'k')   # curva negra

# L√≠neas horizontales (gu√≠as)
for k in range(-2, 3):
    plt.axhline(k*np.pi, color='gray', linestyle='--', linewidth=0.7)

# Estilo del gr√°fico
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("M√©todo de Runge-Kutta 4 para dx/dt = -x/(1+t)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# 4. Using odeint
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definimos la EDO dx/dt = -x/(1+t)
def f(x, t):
    return -x / (1 + t)

# Condiciones iniciales
x0_vals = [-2*np.pi, -np.pi, 0, np.pi, 2*np.pi]

# Intervalo de tiempo
t = np.linspace(0, 10, 300)

# Crear figura
plt.figure(figsize=(6,6))

for x0 in x0_vals:
    x = odeint(f, x0, t)   # resolver con odeint
    plt.plot(t, x, 'k')    # curva negra

# L√≠neas horizontales (gu√≠as)
for k in range(-2, 3):
    plt.axhline(k*np.pi, color='gray', linestyle='--', linewidth=0.7)

# Estilo del gr√°fico
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Soluci√≥n con odeint para dx/dt = -x/(1+t)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# 5. Using solve_ivp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definimos la EDO dx/dt = -x/(1+t)
def f(t, x):
    return -x / (1 + t)

# Condiciones iniciales
x0_vals = [-2*np.pi, -np.pi, 0, np.pi, 2*np.pi]

# Intervalo de tiempo
t_span = (0, 10)
t_eval = np.linspace(0, 10, 300)

# Crear figura
plt.figure(figsize=(6,6))

for x0 in x0_vals:
    sol = solve_ivp(f, t_span, [x0], t_eval=t_eval, method='RK45')  # RK45 por defecto
    plt.plot(sol.t, sol.y[0], 'k')  # curva negra

# L√≠neas horizontales (gu√≠as)
for k in range(-2, 3):
    plt.axhline(k*np.pi, color='gray', linestyle='--', linewidth=0.7)

# Estilo del gr√°fico
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Soluci√≥n con solve_ivp para dx/dt = -x/(1+t)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
